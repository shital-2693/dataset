```python
# import Pandas as pd
import pandas as pd

# create a new data frame
df = pd.DataFrame({'Name': ['1. John peterson  Larter(1994)', '2. Robert Junior paresh(1954)', '3. Jonny jonny Depp(1968)']})

print("Given Dataframe is :\n",df)
```

```python
# splitting the string from fullstop.
s=df['Name'].str.split(pat='.',expand = True)
s
```

```python
#assigning name to column
df['sr_no']= s[0]
df['Movie_Name']= s[1]
```

```python
# dropping the orignal column
df.drop('Name',axis=1,inplace=True)
df
```

```python
#Extracting year from column 
year=df['Movie_Name'].str[-5:-1]
df['Year'] = year
```

```python
df
```

```python
# seperating the name from number 
df['Movie_Name'] = df['Movie_Name'].str.replace(r"\s*\([^()]*\)", "", regex=True).str.strip()
df
```
